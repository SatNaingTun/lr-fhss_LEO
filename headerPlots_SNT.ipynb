```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_base = pd.read_csv("headerResults/snt/100_base.csv")
df_early = pd.read_csv("headerResults/snt/100_early.csv")

df_base.describe()
df_early.describe()

plt.figure()

plt.plot(df_base["sent_packets"],
         df_base["decoded_hrd_pld"],
         label="100 demod - base")

plt.plot(df_early["sent_packets"],
         df_early["decoded_hrd_pld"],
         label="100 demod - early Drop + early Decode")

plt.xscale("log")
plt.xlabel("Sent packets")
plt.ylabel("Total decoded payloads")
plt.legend()
plt.title("Total decoded payloads with 100 demodulators")
plt.show()
```
